```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
size = ((np.random.randn(N) + 1) * 8)**2
colour = np.random.randn(N)
fg,ax = plt.subplots(figsize=(8,4))
l = ax.scatter(x,y,s=size,c=colour)
fg.colorbar(l)
fg.suptitle('Scatter plot')
ax.grid(True)
fg.tight_layout(pad=2)
ax.set_xlim((-0.05,1.05))
ax.set_ylim((-0.06,1.05))
```
