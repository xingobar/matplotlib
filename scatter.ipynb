```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randn(100)
y= x+np.random.randn(100)+10
fig,ax = plt.subplots(figsize=(8,3))
ax.scatter(x,y,color='orchid',alpha=0.5)
ax.grid(True)
fig.suptitle("the example of scatter plot")
fig.tight_layout(pad=2)
```
