```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.array([5,3,4,6])
labels = ['bats', 'cats', 'gnats', 'rats'] 
explode = (0, 0.1, 0, 0) 
colors=['khaki', 'goldenrod', 'tan', 'wheat']
fig,ax = plt.subplots(figsize=(8,3.5))
ax.pie(data,explode=explode,colors=colors,labels=labels,startangle=270,autopct='%1.1f%%')
ax.axis('equal')
fig.suptitle('Pie Chart')
```
